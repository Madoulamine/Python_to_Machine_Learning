## 1. Les Dictionnaires
Contrairement aux listes (indexées par des numéros), les dictionnaires sont indexés par des **clés** (comme un vrai dictionnaire mot/définition). Très rapides et très utilisés en Python.

```python
etudiant = {
    "nom": "Marie",
    "age": 21,
    "filiere": "Informatique"
}

print(etudiant["nom"])

# Ajouter/modifier une entrée
etudiant["moyenne"] = 15.5
print(etudiant)
```

```python
# Création d'un dictionnair représentant de mon joueur
joueur = {
    "nom": "Lamine",
    "niveau": 5,
    "vie": 100,
    "pieces": 250
}

# Afficher le niveau de mon joueur
print("Niveau :", joueur["niveau"])

# Modifier la vie de mon joueur
joueur["vie"] = 80

# Ajouter une nouvelle information
joueur["arme"] = "Épée"

# Afficher toutes les informations du joueur
print(joueur)
```

Je crée un dictionnaire avec plusieurs informations sur un joueur.
Chaque information est stockée sous la forme clé : valeur.
joueur["niveau"] permet d'accéder uniquement à la valeur associée à la clé niveau.
joueur["vie"] = 80 modifie la valeur existante.
joueur["arme"] = "Épée" ajoute une nouvelle clé dans le dictionnaire.
print(joueur) affiche le dictionnaire complet.

## 2. Les Tuples
Comme les listes, mais **immuables** (on ne peut pas les modifier après création). Plus rapides et sécurisent les données.

```python
coordonnees = (48.8566, 2.3522) # Latitude, Longitude de Paris
print("Latitude :", coordonnees[0])
# coordonnees[0] = 50.0  # ERREUR ! Les tuples ne se modifient pas.
```

```python
# Un tuple contenant les informations d'un jeu vidéo
jeu = ("FIFA 25", "Sport", 2025, 69.99)

# 1. Afficher le nom du jeu
print("Nom :", jeu[0])

# 2. Afficher le genre du jeu
print("Genre :", jeu[1])

# 3. Afficher le prix du jeu
print("Prix :", jeu[3], "€")

# 4. Afficher le nombre d'informations contenues dans le tuple
print("Nombre d'informations :", len(jeu))

# 5. Vérifier si l'année 2025 est présente dans le tuple
print(2025 in jeu)

# 6. Parcourir le tuple avec une boucle et afficher chaque élément
print("\nInformations du jeu :")
for information in jeu:
    print(information)

# 7. Essayer de modifier le prix du jeu (Que se passe-t-il ?)
# jeu[3] = 59.99
```

## 3. Les Compréhensions de Listes
Une façon élégante et 'Pythonique' de créer des listes à partir de boucles.

```python
# Version classique
carres = []
for i in range(5):
    carres.append(i ** 2)

# Version compréhension de liste
carres_comp = [i ** 2 for i in range(5)]
print(carres_comp)
```

## 4. Gestion des Exceptions (try/except)
Pour éviter que le programme ne plante, on 'attrape' les erreurs.

```python
try:
    nombre = int("Ce n'est pas un nombre")
except ValueError:
    print("Erreur : vous n'avez pas entré un entier valide.")
finally:
    print("Ce bloc s'exécute dans tous les cas.")
```

```python
try:
    # Demander l'âge à l'utilisateur
    age = int(input("Entrez votre âge : "))

    # Afficher l'âge si la saisie est correcte
    print("Vous avez", age, "ans.")

except ValueError:
    # Ce bloc s'exécute si l'utilisateur n'entre pas un entier
    print("Erreur : veuillez entrer un nombre entier.")

finally:
    # Ce bloc s'exécute dans tous les cas
    print("Merci d'avoir utilisé le programme.")
```

## 🏆 Défi du Jour 4 : Carnet d'Adresses Sécurisé

1. Créez un dictionnaire `carnet` avec deux contacts (ex: "Alice": "0612345678", "Bob": "0698765432").
2. Demandez à l'utilisateur de saisir un nom avec `input()`.
3. Utilisez un bloc `try/except` (ou un `if`) pour gérer le cas où le contact n'existe pas. Si le contact n'existe pas, affichez "Contact introuvable". Sinon, affichez son numéro.
4. *(Bonus)* : Utilisez une compréhension de liste pour créer une liste de tous les noms du carnet qui commencent par la lettre 'A'.

```python
# Écrivez votre code pour le défi ici
# Création du dictionnaire contenant les contacts
carnet = {
    "Alice": "0612345678",
    "Bob": "0698765432",
}

# Demander à l'utilisateur de saisir un nom
nom = input("Entrez le nom du contact : ")

try:
    # Essayer d'afficher le numéro correspondant au nom saisi
    print("Numéro de téléphone est :", carnet[nom])

except KeyError:
    # Ce bloc s'exécute si le nom n'existe pas dans le dictionnaire
    print("Contact introuvable.")

# Bonus : Créer une liste des noms commençant par la lettre 'A'
contacts_A = [contact for contact in carnet if contact.startswith("A")]

# Afficher le résultat
print("Contacts commençant par 'A' :", contacts_A)
```
