## 1. Définir et appeler une fonction
Une fonction est un bloc de code nommé. Elle peut prendre des *paramètres* en entrée et *retourner* un résultat.

```python
def saluer(nom, age):
    """Ceci est une docstring : elle explique ce que fait la fonction."""
    message = "Bonjour " + nom + ", tu as " + str(age) + " ans."
    return message

# Appel de la fonction
resultat = saluer("Bob", 30)
print(resultat)
```

```python
#Une fonction dénommée calcul_ia() qui peut calculer la somme, la difference, le produit et le quotient de 2 entiers N
def calcul_ia(a, b):
    somme = a + b
    difference = a - b
    produit = a * b
    quotient = a / b
    return somme, difference, produit, quotient

#Appel de la fonction
somme, difference, produit, quotient = calcul_ia(10, 5)
print("Somme :", somme)
print("Différence :", difference)
print("Produit :", produit)
print("Quotient :", quotient)
```

## 2. Paramètres par défaut
On peut donner des valeurs par défaut aux paramètres. Ils deviennent optionnels lors de l'appel.

```python
def preparer_cafe(type_cafe="Espresso"):
    return "Voici votre " + type_cafe

print(preparer_cafe())               # Utilise la valeur par défaut
print(preparer_cafe("Cappuccino"))   # Surcharge la valeur par défaut
```

## 3. Portée des variables (Scope)
Les variables créées *à l'intérieur* d'une fonction n'existent pas à l'extérieur. C'est la portée locale.

```python
x = 10  # Variable globale

def ma_fonction():
    x = 5  # Variable locale
    print("x à l'intérieur :", x)

ma_fonction()
print("x à l'extérieur :", x)
```

## 4. Les Modules
Un module est un fichier contenant des fonctions. On l'importe avec `import`.

```python
import math

# On accède aux fonctions du module avec un point
racine = math.sqrt(16)
print("La racine carrée de 16 est", racine)
```

**Calculatrice avec math**

Créez une calculatrice en Python qui permet à l'utilisateur de choisir une opération (addition, soustraction, multiplication, division, racine carrée ou puissance).

**Consignes** :

Créez une fonction pour chaque opération de base.
Importez le module math.
Utilisez math.sqrt() pour la racine carrée et math.pow() pour la puissance.
Affichez le résultat de l'opération choisie.

```python
import math

# Fonctions de base
def addition(a, b):
    return a + b

def soustraction(a, b):
    return a - b

def multiplication(a, b):
    return a * b

def division(a, b):
    if b == 0:
        return "Erreur : division par zéro."
    return a / b

# Menu
print("=== Calculatrice ===")
print("1. Addition")
print("2. Soustraction")
print("3. Multiplication")
print("4. Division")
print("5. Racine carrée")
print("6. Puissance")

choix = int(input("Choisissez une opération (1-6) : "))

if choix in [1, 2, 3, 4, 6]:
    a = float(input("Entrez le premier nombre : "))
    b = float(input("Entrez le deuxième nombre : "))

    if choix == 1:
        print(f"{a} + {b} = {addition(a, b)}")
    elif choix == 2:
        print(f"{a} - {b} = {soustraction(a, b)}")
    elif choix == 3:
        print(f"{a} * {b} = {multiplication(a, b)}")
    elif choix == 4:
        print(f"{a} / {b} = {division(a, b)}")
    elif choix == 6:
        print(f"{a} ^ {b} = {math.pow(a, b)}")

elif choix == 5:
    a = float(input("Entrez un nombre : "))
    if a >= 0:
        print(f"√{a} = {math.sqrt(a)}")
    else:
        print("Erreur : impossible de calculer la racine carrée d'un nombre négatif.")

else:
    print("Choix invalide.")
```
