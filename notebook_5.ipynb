# Jour 5 - 17 Juillet 2026 : Programmation Orientée Objet (POO)

Félicitations pour être arrivé jusqu'ici ! La POO est un paradigme de programmation puissant qui consiste à modéliser le monde réel en créant des **Objets** issus de **Classes**. Vous passez ainsi au niveau intermédiaire.

## 1. Classes et Objets
- Une **Classe** est un plan de fabrication (ex: le concept d'une Voiture).
- Un **Objet** est une instance de ce plan (ex: ma voiture spécifique rouge).
- Les **Attributs** sont les caractéristiques (ex: couleur, marque).
- Les **Méthodes** sont les actions (ex: démarrer, freiner).

```python
class Voiture:
    # La méthode __init__ est le constructeur, appelé à la création de l'objet
    def __init__(self, marque, couleur):
        self.marque = marque    # self représente l'objet en cours de création
        self.couleur = couleur
        self.vitesse = 0        # Attribut par défaut

    def accelerer(self):
        self.vitesse += 10
        print(f"La {self.marque} roule à {self.vitesse} km/h.")

# Création d'objets (instances)
ma_voiture = Voiture("Renault", "Bleue")
ta_voiture = Voiture("Peugeot", "Rouge")

ma_voiture.accelerer()
ma_voiture.accelerer()
```

## 2. L'Héritage
L'héritage permet de créer une nouvelle classe qui réutilise les attributs et méthodes d'une classe existante (la classe parent).

```python
class VoitureElectrique(Voiture):  # Hérite de Voiture
    def __init__(self, marque, couleur, autonomie):
        super().__init__(marque, couleur)  # Appelle le constructeur du parent
        self.autonomie = autonomie

    def recharger(self):
        print("Batterie en charge...")

tesla = VoitureElectrique("Tesla", "Blanche", 500)
tesla.accelerer()  # Héritée de Voiture
tesla.recharger()   # Spécifique à VoitureElectrique
```

## 🏆 Défi du Jour 5 : Modélisation d'un Compte Bancaire

1. Créez une classe `CompteBancaire`.
2. Dans le constructeur `__init__`, initialisez un attribut `solde` à 0 et un attribut `titulaire` (le nom de la personne).
3. Créez une méthode `deposer(montant)` qui augmente le solde.
4. Créez une méthode `retirer(montant)` qui diminue le solde. Utilisez un `try/except` ou une condition `if` pour empêcher le retrait si le solde est insuffisant (affichez "Fonds insuffisants").
5. Instanciez un compte pour "John" avec 0€ au départ. Déposez 100€, essayez de retirer 50€, puis essayez de retirer 200€.

```python
# Écrivez votre code pour le défi final ici
# Définition de la classe CompteBancaire
class CompteBancaire:

    # Constructeur : appelé lors de la création d'un objet
    def __init__(self, titulaire):
        # Nom du titulaire du compte
        self.titulaire = titulaire

        # Solde initial
        self.solde = 0

    # Méthode permettant de déposer de l'argent
    def deposer(self, montant):
        self.solde += montant
        print(f"{montant} € déposés sur le compte de {self.titulaire}.")
        print(f"Nouveau solde : {self.solde} €")

    # Méthode permettant de retirer de l'argent
    def retirer(self, montant):

        # Vérifier si le solde est suffisant
        if montant <= self.solde:
            self.solde -= montant
            print(f"{montant} € retirés.")
            print(f"Nouveau solde : {self.solde} €")
        else:
            print("Fonds insuffisants.")

# Création d'un compte bancaire pour John
compte = CompteBancaire("John")

# Dépôt de 100
compte.deposer(100)

# Retrait de 50
compte.retirer(50)

# Tentative de retrait de 200
compte.retirer(200)
```
